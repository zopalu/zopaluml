# Implement Gradient Descent Algorithm to find the local minima of a function.
For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.


```python
current_x=2
rate=0.1
precision=0.0000001
prev_step_size=1
max_iters=1000
iters=0

df = lambda x: 2*(x+3)
```

```python
while prev_step_size > precision and iters < max_iters:
    prev_x = current_x
    current_x = current_x - rate * df(prev_x)
    prev_step_size = abs(current_x - prev_x)
    iters = iters + 1
    print("Iteration",iters,"\nX value is",current_x) #Print iterations
    
print("Local minima occurs as: ", current_x)
```
